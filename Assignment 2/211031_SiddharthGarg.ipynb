```python
unique_surname = database['VoterLastNameEn'].unique()
a = []
for i in unique_surname:
  count =  (database['VoterLastNameEn'] == i).sum()
  a.append(count)

x = unique_surname
y = a
y = np.array(y)

surname_analysis = pd.DataFrame({"Surname":x, "Count":y})
print(surname_analysis)
```

```python
religion = np.array(['Shaikh', 'Pathan', 'Momin', 'Bohra' , 'Ghanchi', 'Malik', 'Garasia', 'Syed', 'Fakir', 'Musalman', 'Mansoori', 'Khatki', 'Chhippa', 'Kadia', 'Sipahi', 'Tai', 'Memon', 'Patel Vohra', 'Sindhi','Sandhi', 'Hajam', 'Pinjara', 'Mughal', 'Kasbati', 'Kumbhar', 'Dudhwala', 'Baluch', 'Makrani', 'Behlim', 'Qureshi', 'Mirasi', 'Khatri', 'Khokhar', 'Dhobi', 'Jat'])
muslim_religion = []
x = []
for i in religion:
  count =  (database['VoterLastNameEn'] == i).sum() + (database['VoterFirstNameEn'] == i).sum()
  if(count != 0):
    muslim_religion.append(count)
    x.append(i)
    
muslim_religion = np.array(muslim_religion)
x = np.array(x)
print(x)
religion_analysis = pd.DataFrame({"Muslim surname":x, "Count":muslim_religion})
print(religion_analysis)
```
